```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('../AmesHousing.csv')


numeric_df = df.select_dtypes(include=['number']).drop(columns=['Order', 'PID'], errors='ignore')


correlations = numeric_df.corr()['SalePrice'].sort_values(ascending=False)


top_features = correlations.head(15).index
top_corr_matrix = numeric_df[top_features].corr()


sns.set_theme(style="white")

sns.heatmap(top_corr_matrix, annot=True, cmap='RdBu_r', center=0, fmt=".2f")
plt.title('Features with the Strongest Relationship with SalePrice')
plt.savefig('correlation_heatmap.png')

print(correlations.head(15))
```
